**Q8. _\[5 points\]_** From the flights dataframe, create another column called "penalty" that shows the penalty that the airline has to pay back to customer. If the arrival delay is less than 10 minutes, the penalty is 0. If the arrival delay is greater than 10 minutes, the penalty is $2 per minute.

```python
#Q8. Answer
def f(flights):
    if flights['arr_delay'] < 10:
        val = 0
    else :
        val = flights['arr_delay']  * 2
    return val
flights['penalty'] = flights.apply(f, axis=1)
flights
```

**Q9. _\[5 points\]_** From the flights dataframe, if the departure delay is negative, it means the plane leaves earlier than the schedule. If the arrival delay is positive, it means the flight is late. If the total delay is calculated as arr_delay - dept_delay. Create a dataframe that shows the maximum total delay of each carrier (aka: airline).

```python
#Q9. Answer
max_delay = flights
max_delay["total_delay"] = max_delay["arr_delay"] - max_delay["dep_delay"]
final = max_delay.groupby('carrier').total_delay.max()
final.reset_index()
```
