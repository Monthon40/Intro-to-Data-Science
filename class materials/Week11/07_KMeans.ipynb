# Now, time for K Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data4c[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

## For real data, it ends here. Since we never have labels of our data. 
Anyway, here we create the data ourselves. So, we can try to see how good it is.

```python
# Visualize the result, let's add the class to our df
df4c['pred_class']=kmeans.labels_
```

```python
# Let's plot the redictions
sns.scatterplot(data=df4c,x='feature1',y='feature2',size="pred_class", palette='colorblind')
```

```python
# Let's see the actual class
sns.scatterplot(data=df4c,x='feature1',y='feature2',size="the_class", palette='colorblind')
```

```python
# Since K Means may assig any label (number) to a cluster, to visualize them, let's change the number
def remap(x):
    if x==0: return 2
    if x==1: return 0
    if x==2: return 1
    if x==3: return 3

df4c['pred_class_remap'] = df4c['pred_class'].apply(remap)
```

```python
# Now, we can calculate the difference between the prediction and the data
df4c['diff_class']=df4c['the_class']-df4c['pred_class_remap']
```

```python
# If they are the same, mismatched will be marked as 0, else 1
df4c['mismatched'] = df4c.apply(lambda x: 1 if x.diff_class != 0 else 0, axis=1)
```

```python
df4c.head(10)
```

```python
# Let's see which one they got it wrong
sns.scatterplot(data=df4c,x='feature1',y='feature2',hue="mismatched", palette='colorblind')
```

# What should be the number of cluster?
Elbow Method

```python
SSE = list()
```

```python
kmeans2c = KMeans(n_clusters=2)
kmeans2c.fit(data4c[0])
SSE.append(kmeans2c.inertia_)
```

```python
kmeans3c = KMeans(n_clusters=3)
kmeans3c.fit(data4c[0])
SSE.append(kmeans3c.inertia_)
```

```python
kmeans4c = KMeans(n_clusters=4)
kmeans4c.fit(data4c[0])
SSE.append(kmeans4c.inertia_)
```

```python
kmeans5c = KMeans(n_clusters=5)
kmeans5c.fit(data4c[0])
SSE.append(kmeans5c.inertia_)
```

```python
kmeans6c = KMeans(n_clusters=6)
kmeans6c.fit(data4c[0])
SSE.append(kmeans6c.inertia_)
```

```python
sse_df = pd.DataFrame(SSE,columns=['y'])
sse_df['x']= range(2,7)
```

```python
sns.scatterplot(data=sse_df,x='x',y='y')
```
