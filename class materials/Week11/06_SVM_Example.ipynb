# ICA -  SVM
Class a support vector classifier (SVC) to classify a  species of the iris dataset. Measure and report the performance


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head(3)
```

```python
sns.pairplot(iris, hue='species')
```

```python
setosa = iris[iris['species']=='setosa']
sns.kdeplot(setosa.sepal_width,setosa.sepal_length,cmap='plasma',shade=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = iris.drop('species',axis=1)
```

```python
y = iris['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=161)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
