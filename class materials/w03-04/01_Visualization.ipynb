### Subplot 
We can make multiple plots in one figure by using subplot(s) functions. (Actually, the official tutorial show this way first)
 

````python
plt.subplot(1,2,1)  # (#row, #col, #subplot_index)
plt.plot(x,y,'b')

plt.subplot(1,2,2)  # (#row, #col, #subplot_index)
plt.plot(x2,y2,'r')
````

Let's look at the example from the official tutorial.

````python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
````

From the code above, there are many things going on in the first line of code. It calls plt.subplots(). The function returns a **figure** and an **axes**.

The second line, it plots the curve out.

We can try to understand what do they mean by reconstruction step by step.

**figure**

Figure is a blank canvas. We can create it by calling

```fig = plt.figure()```

**axes**
>The Axes contains most of the figure elements: Axis, Tick, Line2D, Text, Polygon, etc., and sets the coordinate system. -- matplotlib.org


````python
my_fig = plt.figure() #create a figure object

axes = my_fig.add_axes([0.2,0.2,0.6,0.6]) #add an axes [Left,Buttom,Width,Height]
````

````python
my_fig1 = plt.figure()

axes = my_fig1.add_axes([0.1,0.1,0.8,0.8]) #add an axes [Left,Buttom,Width,Height]

axes.plot(x,y)

axes.set_xlabel('X value')
axes.set_ylabel('Y value')
axes.set_title('A parabora plot')
````

````python
axes.set_title('Ha Ha Ha')
my_fig1
````

## Let's do something more complicated to see how do we do the subplots

````python
my_fig2 = plt.figure()

axes1 = my_fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = my_fig2.add_axes([0.2,0.2,0.3,0.6])
````

````python
axes1.set_title("Ha Ha")
axes2.set_title("Ho Ho")
my_fig2
````

````python
axes1.plot(x2,y2)
axes2.plot(x,y)
my_fig2
````

You can start to see that we can overlay many things to the figure. This allows us to create something complicated.
