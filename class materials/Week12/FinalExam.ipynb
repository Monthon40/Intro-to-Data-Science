## Q7 \[15 points\]
Given the data in **blackbox.csv**, use K Means to analyze that how many clusters of data points are in this file. Provide evidence to support your answer.

````python
blackbox = pd.read_csv("./datasets/dummy.csv")
# blackbox
````

````python
from sklearn.cluster import KMeans
SSE = list()
X = blackbox[['feature1', 'feature2', 'feature3', 'feature4']]
````

````python
kmeans2c = KMeans(n_clusters=2)
kmeans2c.fit(X)
SSE.append(kmeans2c.inertia_)
````

````python
kmeans3c = KMeans(n_clusters=3)
kmeans3c.fit(X)
SSE.append(kmeans3c.inertia_)
````

````python
kmeans4c = KMeans(n_clusters=4)
kmeans4c.fit(X)
SSE.append(kmeans4c.inertia_)
````

````python
kmeans5c = KMeans(n_clusters=5)
kmeans5c.fit(X)
SSE.append(kmeans5c.inertia_)
````

````python
kmeans6c = KMeans(n_clusters=6)
kmeans6c.fit(X)
SSE.append(kmeans6c.inertia_)
````

````python
kmeans7c = KMeans(n_clusters=7)
kmeans7c.fit(X)
SSE.append(kmeans7c.inertia_)
````

````python
kmeans8c = KMeans(n_clusters=8)
kmeans8c.fit(X)
SSE.append(kmeans8c.inertia_)
````

````python
kmeans9c = KMeans(n_clusters=9)
kmeans9c.fit(X)
SSE.append(kmeans9c.inertia_)
````

````python
sse_df = pd.DataFrame(SSE,columns=['y'])
sse_df['x']= range(2,10)# number of K coressponding to that SSE
````

````python
plt.figure(figsize=(10,6))
plt.plot(sse_df["x"],sse_df["y"],color="blue",linestyle="--",marker="o",markerfacecolor="red",markersize=10)
plt.xlabel("K")
plt.ylabel("SSE")
````
